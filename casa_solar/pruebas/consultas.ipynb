```python
import mysql.connector
import pandas as pd
from IPython.display import display
pd.options.display.max_columns = None
 
connection = None

connection = mysql.connector.connect(
    host='localhost', user='root',passwd='admin', database='casasolar_mayo')

#SELECT
def select_all_table(table_name, limit, connection, random):
    if random==1:
        rand='ORDER BY RAND()'
    else:
        rand=''
    q = f"""
    SELECT * FROM {table_name} {rand} LIMIT {limit} ;
    """
    df = pd.read_sql_query(q, connection)
    print('Consulta SELECT exitosa')        
    return df
```

```python
select_all_table('hito', 100, connection, random=0)
```

```python
select_all_table('postulacion', 3, connection, random=1)
```
